(1) Create a 10x10 grid to visualize 10 examples of each digit. Python hints:

- plt.rc() for setting the colormap, for example to black and white
- plt.subplot() for creating subplots
- plt.imshow() for rendering a matrix
- np.array.reshape() for reshaping a 1D feature vector into a 2D matrix (for rendering)

```python
def P1(num_examples=10):

    ### STUDENT START ###
    # num_digits is the number of digits that exist in the dataset that you would like to display
    num_digits = 10
    # Create a list called doneList for checking whether you are done collecting samples for each digit.
    # Initialize this list to FALSE for all values.  Set a value to TRUE if it's index is equal to digit
    # and you have collected num_examples of that digit.
    doneList = []
    # exampleLists is a list of lists.  It will contain num_digits lists.  Each of those lists will be a
    # list that has num_examples elements.  Each of these elements will be a digit image.  Initialize
    # this variable to a list of empty lists.
    exampleLists = []
    for i in range(num_digits):
        doneList.append(False)
        exampleLists.append([])
    
    # As long as doneList does not contain all True values keep iterating.  When it contains all True values,
    # then num_examples will have been collected for each digit.
    i=0
    while False in doneList:
        # Set curDigit to the target value of the image that is being examined
        curDigit = int(Y[i])
        # If the list for the curDigit does not have num_examples examples in it then append the current 
        # image to it, otherwise set that digits doneList value to True.
        if len(exampleLists[curDigit]) < num_examples:
            exampleLists[curDigit].append(X[i])
        else:
            doneList[curDigit] = True
        i += 1

    # Initialize a new plot.
    plt.figure(figsize=(10,10))

    # Loop throught the exampleLists variable and plot all of the images
    for i in range(num_digits):
        for j in range(num_examples):
            # reshape the image to a 28x28 2D array
            xMatrix = exampleLists[i][j].reshape((28,28))
            # Create a subplot space that is num_digits high and num_examples wide
            # Calculate the location of the image of this iteration based on i, j and num_examples
            ax = plt.subplot(num_digits, num_examples, ((num_examples*i)+j)+1)
            # Turn off tick marks to keep things clean.
            plt.setp(ax, xticks=(), yticks=())
            # render the matrix, set the color map to Greys
            plt.imshow(xMatrix, cmap="Greys")
    
    # Render the plots.
    plt.show()    
    ### STUDENT END ###

P1(10)
```

- time.time() gives a wall clock value you can use for timing operations

```python
def P3(train_sizes, accuracies):

### STUDENT START ###

    predict = {}
    times = {}
    neigh = KNeighborsClassifier(n_neighbors=1)
    for t in train_sizes:
        # ??should this be here or after neigh call
        time_start = time.time()
        neigh.fit(train_data[:t], train_labels[:t])
        time_end = time.time()
        times[t] = time_end - time_start
        predict[t] = neigh.predict(dev_data[:t])
        score = neigh.score(dev_data[:t], dev_labels[:t])
        accuracies.append(score)
        print "Classification Report for t= " + str(t)
        print classification_report(dev_labels[:t], predict[t], digits=4)
        print "Run time= " + str(times[t])
        print "Score= " + str(score)


### STUDENT END ###

train_sizes = [100, 200, 400, 800, 1600, 3200, 6400, 12800, 25000]
accuracies = []
P3(train_sizes, accuracies)
```
