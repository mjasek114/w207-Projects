(6) A common image processing technique is to smooth an image by blurring. The idea is that the value of a particular pixel is estimated as the weighted combination of the original value and the values around it. Typically, the blurring is Gaussian -- that is, the weight of a pixel's influence is determined by a Gaussian function over the distance to the relevant pixel.

Implement a simplified Gaussian blur by just using the 8 neighboring pixels: the smoothed value of a pixel is a weighted combination of the original value and the 8 neighboring values. Try applying your blur filter in 3 ways:
- preprocess the training data but not the dev data
- preprocess the dev data but not the training data
- preprocess both training and dev data

Note that there are Guassian blur filters available, for example in scipy.ndimage.filters. You're welcome to experiment with those, but you are likely to get the best results with the simplified version I described above.

```python
def P6():
    
### STUDENT START ###
    def blur_filter(d):
        # Takes a dataset d and returns the blurred version of d.  Blurring is accomplished by
        # taking a weighted combination of the original value and the 8 neighboring values.
        
        # Set the weights.  w is the weight for the original value.  the other weights are labeled
        # by the directional abbreviations for north N, south S and so on according to what direction
        # they are from the perspective of the original point
        w = .4
        wN, wNE, wE, wSE, wS, wSW, wW, wNW = 1, 1, 1, 1, 1, 1, 1, 1
        
        # Create a new empty dataset with the same shape as d
        new_d = np.empty(d.shape)
        # Loop through d and set the values of the new_d based on the blurring formula of adding
        # up the nearest neighbors + the original value times each values respective weight and then
        # dividing by the sum of the weights.
        # i1 will represent which image the program is working on.  i2 will represent which pixel value
        # the program is working on.  The pixel values go from 0 to 783 in a 28x28 matrix.
        for i1 in range(len(d)):
            for i2 in range(len(d[i1])):
                # if the value is in the top row of the image matrix, then it's a special case.  It does not
                # have any top neighbors.
                if (0 <= i2 and i2 <= 27):
                    if i2 == 0:
                        value = (w*d[i1][i2]+wE*d[i1][i2+1]+wS*d[i1][28]+wSE*d[i1][29])/(w+wE+wS+wSE)
                    elif i2 == 27:
                        value = (w*d[i1][i2]+wW*d[i1][i2-1]+wSW*d[i1][54]+wS*d[i1][55])/(w+wW+wSW+wS)
                    else:
                        value = (w*d[i1][i2]+wW*d[i1][i2-1]+wE*d[i1][i2+1]+wSW*d[i1][i2+27] \
                                 +wS*d[i1][i2+28]+wSE*d[i1][i2+29])/(w+wW+wE+wSW+wS+wSE)       
                # if the value is in the bottom row of the image matrix, then it's a special case.  It does not
                # have any bottom neighbors.
                elif (756 <= i2 and i2 < 784):
                    if i2 == 756:
                        value = (w*d[i1][i2]+wE*d[i1][i2+1]+wN*d[i1][728]+wNE*d[i1][729])/(w+wE+wN+wNE)
                    elif i2 == 783:
                        value = (w*d[i1][i2]+wW*d[i1][i2-1]+wNW*d[i1][754]+wN*d[i1][755])/(w+wW+wNW+wN)
                    else:
                        value = (w*d[i1][i2]+wW*d[i1][i2-1]+wE*d[i1][i2+1]+wNE*d[i1][i2-27] \
                                 +wN*d[i1][i2-28]+wNW*d[i1][i2-29])/(w+wW+wE+wNE+wN+wNW)                
                # if the value is in the middle of the image matrix, then follow this general formula
                else:
                    if (i2 % 28 == 0):
                        value = (w*d[i1][i2]+wE*d[i1][i2+1]+wNE*d[i1][i2-27]+wN*d[i1][i2-28] \
                                 +wS*d[i1][i2+28]+wSE*d[i1][i2+29])/(w+wE+wNE+wN+wS+wSE)
                    elif ((i2+1) % 28 == 0):
                        value = (w*d[i1][i2]+wW*d[i1][i2-1]+wNW*d[i1][i2-29]+wN*d[i1][i2-28] \
                                 +wS*d[i1][i2+28]+wSW*d[i1][i2+27])/(w+wW+wN+wNW+wSW+wS)
                    else:
                        value = (w*d[i1][i2]+wW*d[i1][i2-1]+wE*d[i1][i2+1]+wSW*d[i1][i2+27] \
                                 +wS*d[i1][i2+28]+wSE*d[i1][i2+29]+wNE*d[i1][i2-27]+wN*d[i1][i2-28] \
                                 +wNW*d[i1][i2-29])/(w+wW+wE+wNE+wN+wNW+wSW+wS+wSE)
                new_d[i1][i2] = value
        return new_d
    
    blurred_dev_data = blur_filter(dev_data)
    blurred_mini_train_data = blur_filter(mini_train_data)
    
    # set the value for k for the K-NN classifier
    k = 1
    # Fit a model and print the accuracy scrore for the regular training data and the regular dev data
    neigh = KNeighborsClassifier(n_neighbors=k)
    neigh.fit(mini_train_data, mini_train_labels) 
    score = neigh.score(dev_data, dev_labels)
    print "Reg Train, Reg Dev Score= " + str(score)
    
    # Fit a model and print the accuracy scrore for the blurred training data and the regular dev data
    neigh.fit(blurred_mini_train_data, mini_train_labels) 
    score = neigh.score(dev_data, dev_labels)
    print "Blurred Train, Reg Dev Score= " + str(score)
    
    # Fit a model and print the accuracy scrore for the regular training data and the blurred dev data
    neigh.fit(mini_train_data, mini_train_labels) 
    score = neigh.score(blurred_dev_data, dev_labels)
    print "Reg Train, Blurred Dev Score= " + str(score)
    
    # Fit a model and print the accuracy scrore for the blurred training data and the blurred dev data
    neigh.fit(blurred_mini_train_data, mini_train_labels) 
    score = neigh.score(blurred_dev_data, dev_labels)
    print "Blurred Train, Blurred Dev Score= " + str(score)
    
    # set the number of blurred examples that the program will print
    num_examples = 5
    
    # Initialize a new plot.
    plt.figure(figsize=(10,4))
    # Loop through the blurred_dev_data and display num_examples examples
    # for each example, first display the regular image, then display the blurred image
    for i in range(num_examples):
        # Reshape the image to a 28x28 2D array
        xMatrix1 = dev_data[i].reshape((28,28))
        # Create a subplot space that is 2 high and 3 wide, calculate the location of the image of 
        # this iteration based on i
        ax1 = plt.subplot(2, num_examples, i+1)
        # Turn off tick marks to keep things clean.
        plt.setp(ax1, xticks=(), yticks=())
        # Render the matrix, set the color map to Greys
        plt.imshow(xMatrix1, cmap="Greys")

        # Reshape the image to a 28x28 2D array
        xMatrix2 = blurred_dev_data[i].reshape((28,28))
        # Create a subplot space that is 2 high and 3 wide, calculate the location of the image of 
        # this iteration based on i
        ax2 = plt.subplot(2, num_examples, i+1+num_examples)
        # Turn off tick marks to keep things clean.
        plt.setp(ax2, xticks=(), yticks=())
        # Render the matrix, set the color map to Greys
        plt.imshow(xMatrix2, cmap="Greys")
        
    # Render the plots.
    plt.show()

### STUDENT END ###

P6()
```

ANSWER:  Below are the accuracy scores for a 1NN classifer using the specified dataset using a weight of .4 for the original value and 1 for all of the neighbors:
    Regular mini_training data, regular dev data:  0.888

(10) Because Naive Bayes is a generative model, we can use the trained model to generate digits. Train a BernoulliNB model and then generate a 10x20 grid with 20 examples of each digit. Because you're using a Bernoulli model, each pixel output will be either 0 or 1. How do the generated digits compare to the training digits?

- You can use np.random.rand() to generate random numbers from a uniform distribution
- The estimated probability of each pixel is stored in feature\_log\_prob\_. You'll need to use np.exp() to convert a log probability back to a probability.

```python
def P10(num_examples):

### STUDENT START ###
    # BernoulliNB(alpha=1.0, binarize=0.0, class_prior=None, fit_prior=True)
    # Train a BernoulliNB model.  Use binarize=0.5, so that all of the values are either
    # 0 or 1.
    bernNB = BernoulliNB(binarize=0.5)
    bernNB.fit(mini_train_data, mini_train_labels)
    
    # Generate num_examples examples of each of the digits
    num_digits = 10
    # Create a list of lists to hold all of the examples
    examples = []
    for i in range(num_digits):
        examples.append([])
        for j in range(num_examples):
            # create and empty array for the shape of the feature probabilities
            d = np.empty(np.exp(bernNB.feature_log_prob_)[i].shape)
            # copy the feature probabilities for digit i to this new array
            np.copyto(d, np.exp(bernNB.feature_log_prob_)[i])
            # create an array of random picks of floats between 0 and 1.  The
            # shape of this array is the same size as array d
            rand = np.random.rand(784)
            # compare the generated random numbers to the probabilities output by the model
            # the probability represents the probability that the pixel should be a 1.
            # if the random number is less than the probability, then that value should
            # be set to 1.  This should have the effect of generating a 1 value a number of
            # times that is consistent with its probability.
            d1 = np.less(rand, d)
            examples[i].append(d1)
  
    # Initialize a new plot.
    plt.figure(figsize=(16,22))

    for i in range(num_digits):
        for j in range(num_examples):
            # reshape the image to a 28x28 2D array
            xMatrix = examples[i][j].reshape((28,28))
            # Create a subplot space that is num_digits wide and num_examples high
            # Calculate the location of the image of this iteration based on i, j and num_digits
            ax = plt.subplot(num_examples, num_digits,((num_digits*j)+i)+1)
            # Turn off tick marks to keep things clean.
            plt.setp(ax, xticks=(), yticks=())
            # render the matrix, set the color map to Greys
            plt.imshow(xMatrix, cmap="Greys")
    
    # Render the plots.
    plt.show()  

### STUDENT END ###

P10(20)
```

ANSWER:  How do the generated digits compare to the training digits?  The generated digits are grainy and it is difficult to tell what each number is.  This is because of the assumption that the features are independent when in reality they are not.  In reality, when you are drawing a particular number the fact that you filled in the pixel before it, does make a difference as to whether you fill in the currect pixel.  This fact is not captured in these generated pixels, so they look a little less real.

(11) Remember that a strongly calibrated classifier is rougly 90% accurate when the posterior probability of the predicted class is 0.9. A weakly calibrated classifier is more accurate when the posterior is 90% than when it is 80%. A poorly calibrated classifier has no positive correlation between posterior and accuracy.

Train a BernoulliNB model with a reasonable alpha value. For each posterior bucket (think of a bin in a histogram), you want to estimate the classifier's accuracy. So for each prediction, find the bucket the maximum posterior belongs to and update the "correct" and "total" counters.

How would you characterize the calibration for the Naive Bayes model?

```python
def P11(buckets, correct, total):
    
### STUDENT START ###
       
    # BernoulliNB(alpha=1.0, binarize=0.0, class_prior=None, fit_prior=True)
    # Train a BernoulliNB model.  Use binarize=0.5, so that all of the values are either
    # 0 or 1.  Set alpha = 0.01 which was found to be the best alpha.
    bernNB = BernoulliNB(alpha=0.01, binarize=0.5)
    # Fit the model with the mini_train_data
    bernNB.fit(mini_train_data, mini_train_labels)
    # Print the accuracy score with the dev data
    score = bernNB.score(dev_data, dev_labels)
    print "BernNB Score= " + str(score)

    # For each prediction, find the maximum posterior probability from the predict_proba list
    # Find the bucket to which this maximum probability belongs.  Update the total and correct
    # counts for that bucket.  Update the total every time, but only update the correct counter
    # if the predicted value is equal to the actual value.
    for i in range(len(bernNB.predict(dev_data))):
        max_prob = max(bernNB.predict_proba(dev_data)[i])
        for j in range(len(buckets)):
            if max_prob <= buckets[j]:
                cur_bucket=j
                break
        total[cur_bucket] += 1
        if bernNB.predict(dev_data)[i] == dev_labels[i]:
            correct[cur_bucket] += 1.0
                
### STUDENT END ###

buckets = [0.5, 0.9, 0.999, 0.99999, 0.9999999, 0.999999999, 0.99999999999, 0.9999999999999, 1.0]
correct = [0 for i in buckets]
total = [0 for i in buckets]

P11(buckets, correct, total)

for i in range(len(buckets)):
    accuracy = 0.0
    if (total[i] > 0): accuracy = correct[i] / total[i]
    print 'p(pred) <= %.13f    total = %3d    accuracy = %.3f' %(buckets[i], total[i], accuracy)
```

ANSWER:  I would characterize the calibration for the Naive Bayes model as 'weakly calibrated' because the accuracies increase as the posterior probability bucket values increase in magnitude.
